Você trabalha em um projeto de análise de dados e precisa adquirir informações de uma página da Wikipedia sobre um tema específico para sua análise. \
No entando, a Wikepidia não fornece uma API direta para acessar os dados, então você decide criar um script Python para automatizar o processo da aquisição de dados. \
Você planeka usar Selenium para navegar até a página da Wikepedia relacionada ao seu tema de interesse, salvar o conteúdo HTML da página, extrair a tabela relevante usando BeautifulSoup e, finalmente processar os dados da tabela com pandas para análise posterior.

```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

url = "https://pt.wikepedia.org/wiki/Wikip%C3%A9dia:P%C3%A1gina_principal"
tema = "selecao brasileira"
#driver = river = webdriver.Chrome(options=chrome_options)

driver = webdriver.Chrome()

#Fazer pesquisa
driver.get(url)
pesquisa = driver.find_element(By.NAME, "search")
pesquisa.send_keys(tema)
pesquisa.submit()


#Tratar os tipos de pesquisar e acessar as paginas wiki
if driver.find_elements(By.XPATH,'//*[@id="disambig"]/table/tbody/tr/td[1]/span/a/img'):
  driver.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[1]/ul/li[1]/a').click()
  print('Pagina desambiguacao')

elif driver.find_elements(By.XPATH,'//*[@id="firstHeading"]'):
  if driver.find_elements(By.XPATH,'//*[@id="firstHeading"]')[0].text == 'Resultados da pesquisa':
    print('Pesquisa avançada')
    driver.find_element(By.XPATH,'//*[@id="mw-content-text"]/div[3]/div[4]/ul/li[1]/div[2]/div[2]/div[1]/a').click()
  else:
    print('Pagina da wiki')

html = driver.page_source
driver.quit()
```

```python
soup = BeautifulSoup(html, 'html.parser')
tabelas = soup.find_all('table')
for tabela in tabelas:
  df = pd.read_html(str(tabela))[0]
  print(df.head())
```
